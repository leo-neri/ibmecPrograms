# Concorrência monopolista

- É um híbrido entre concorrência perfeita e monopólio
    - Monopólio por conta da diferenciação do produto
    - Concorrência perfeita por conta das similaridades com outros produtos
- Produtos diferenciados entre as empresas
- Substitutos próximos (mas não perfeitas)
- Livre entrada e saída do mercado
    - Implica que quanto mais lucro (econômico), mais empresas no mercado
    - E quanto mais prejuízo, menos empresas também
