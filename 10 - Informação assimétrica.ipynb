# Informação assimétrica

- Ocorre quando um dos agentes do mercado sabem mais de algo do que outros

## Tipos

### Informação sobre as características dos agentes

- 1. Qualidade do produto
    - Normalmente, o vendedor conhece melhor a qualidade do produto do que o comprador
- 2. Um recém-formado sabe mais sobre suas habilidades do que as empresas que estão contratando
- 3. Uma empresa que precisa de crédito para expandir seu negócio sabe mais sobre os riscos do negócio que os bancos a quem ele vai pedir empréstimo
- 4. A pessoa que compra um segurp-saúde sabe mais sobre sua condição de saúde que a empresa seguradora
- 5. A companhia aérea não sabe se o passageiro é um executivo ou um turista (a companhis gostaria de cobrar mais caro do turista)<br>
#### Nos exemplos 1 e 2, o ofertante tem mais informação que o demandante
#### Nos exemplos 3 a 5, o demandante tem mais informação que o ofertante

#### Seleção adversa

- <strong>Seleção adversa</strong> é um problema que pode surgir quando há assimetria de informação nas <strong>características dos agentes</strong>
- Seleção adversa ocorre quando a assimetria faz os <strong>melhores agentes ou produtos deixarem o mercado</strong>, ficando apenas os piores
- Por exemplo, quando a empresa toma uma medida que faz apenas os melhores trabalhadores pedirem demissão

#### Soluções

- No casos em que o ofertante não sabe quais as características dos demandantes, ele cria <strong>diferentes opções de pacotes de produto</strong>, sabendo que cada tipo de demandante irá optar por um pacote diferente
- Por exemplo, quando o banco vai realizar um empréstimo para uma empresa, ele pode oferecer duas possibilidades de contrato, um contrato de juro alto e garantias baixas, ou um contrato de juro baixo mas garantias altas
- Outra possibilidade ocorre quando o demandante tenta  <strong>sinalizar qual o seu tipo</strong> para o ofertante
- Da mesma forma, o ofertante de um produto de qualidade também pode tentar sinalizar isso para o demandante

### Informação assimétrica sobre as ações dos agentes

- 1. O acionista que uma empresa não observa completamente como seus executivos estão gerindo a empresa
- 2. O dono da empresa não observa completamente se seus funcionários estão se esforçando ou não
- 3. A empresa de segure de carro não observa se seus segurados dirigem bem ou não, ou quais cuidados eles estão tomando para evitar o roubo do carro
- 4. Uma empresa de seguro de saúde não observa se seus clientes adotam ou não algum comportamento de risco (tipo, se resolvem praticar algum esporte radical)

#### Risco moral

- <strong>Risco moral</strong> é um problema que surge quando há assimetria de informação nos <strong>comportamentos dos agentes</strong>
- Risco moral é a situação em que o <strong>agente que não está sendo observado/monitorado passa a se comportar de forma inadequada</strong> (da perspectiva do outro agente)
- Por exemplo, quando o executivo da empresa passa a tomar decisões em benefício próprio, contrários aos interesses dos acionistas

#### Soluções

- Quando para um dos lados de um relacionamento é custoso observar se o outro está se comportando bem, há algumas soluções
- Uma é condicionar um maior pagamento a um maior desempenho
- Por exemplo, o dono da loja paga uma comissão sobre as vendas para os vendedores
- No caso da seguradora de saúde, uma forma de tentar fazer a pessoa ser cuidadosa com a própria saúde é não cobrir totalmente as despesas médicas, para obrigar o cliente a arcar com parte dos gastos (<i>put your skin in the game</i>)

## Risco moral

- O termo <strong>risco moral</strong> vem do mercado de <strong>seguros</strong>
- Se refere ao incentivo que o segurado tem de ser menos cuidadoso após adquirir o seguro
- Por extensão, acabou se referindo a qualquer situação em que há incentivos para o agente "se comportar mal" quando suas ações não são perfeitamente observadas
- Dizemos que risco moral é um problema gerado pela situação em que há uma assimetria de informação: um agente não observa com precisão o comportamento de outro agente
- Risco moral traz problemas para o funcionamento dos mercados, ou seja, gera ineficiências na alocação dos recursos

### O prêmio justo no mercado de seguros

- Suponha o mercado de seguro contra roubo de carro
- O valor de cada carro é \$1000\$ e há uma probabilidade de 5\% de roubo
- Ignorando os custos operacionais da seguradora, qual o preço que ela deveria cobrar para vender um seguro completo contra toubo?
- Um seguro é completo quando ele segura 100\% do valor do bem segurado
- O preço do seguro é chamado de <strong>prêmio do seguro</strong>
- Em um mercado perfeitamente competitivo, o lucro econômico esperado da seguradora é 0
- O premio que gera lucro zero é chamado de <strong>prêmio justo</strong>
- No exemplo, o prêmio justo é \\$50 $(\$1000$ x $0.05)$, pois é o valor médio que a seguradora irá pagar por carro segurado
- Assim, a $50, o lucro econômico da seguradora é 0
- Mas suponha que o segurado, após fazer o seguro, se torne menos cuidadoso: passa a não se proucpar com o local em que ele estaciona ou deixe de gastar com equipamentos de segurança (alrme, trava, etc.)
- Nesse caso, a probabilidade de roubo aumenta e o prêmio de \\$50\\$ faz a seguradora ter prejuízo
- A seguradora, então, precisa cobrar mais que \\\$50
- Mas, ao fazer isso, pela curva de demanda ser negativamente inclinada, a quantidade de seguros negociados diminui
- Resumindo: o "mau" comportamento dos segurados aumenta a probabilidade de roubo, aumenta o prêmio do seguro e reduz o número de carros segurados
- Outra possibilidade é a seguradora não vender mais o seguro completo
- No seguro completo, a seguradora não cobre 100% do valor do carro roubado
- Assim, a seguradora transfere parte do risco para o segurado
- Isso aumenta o incentivo para que a pessoa se comporte de forma mais cuidadosa
- Mesmo assim, há uma ineficiência: as pessoas gostariam de comprar o seguro completo e as firmas gostariam de vender o seguro completo
- Mas as firmas não vendem o seguro completo porque sabem que, ao fazêlo, os consumidores ficarão menos cuidadosos

### O mercado do seguro de saúde

- No caso do seguro saúde, o mesmo pode acontecer
- Ao ficar segurada, uma pessoa pode passar a ter um comportamento menos saudável e mais arriscado (por exemplo, passa a praticar esportes radicais)
- Novamente, as seguradoras de saúde tentam reduzir esse problema não cobrindo 100% dos gastos médicos dos clientes
- Assim, parte das despesas médicas terão que ser pagas pelo próprio segurado
- Ao não cobrir 100% do valor do carro ou 100% do tratamento médico, as seguradoras estão transferindo parte do risco para o segurado
- Diz-se que elas estão fazendo o segurado "colocar a própria pele no jogo" (<i>putting your skin in the game</i>)

### Risco moral do mercado de trabalho

- Há várias instâncias de "mau comportamento" no mercado de trabalho
- Por exemplo, o emprego pode roubar a empresa
- Um caso menos grave é aquele em que o trabalhador se esforça pouco
- Outro exemplo se refere ao comportamento do executibo de uma companhia que toma decisões que são mais em interesse próprio que do interesse dos acionistas
- Por exemplo, o executivo compra um jato para a companhia, mas quem irá usar será ele mesmo: será que ele comporu apenas para o próprio conforto ou realmente irá trazer mais lucros para a companhia?

### O problema do agente e do principal

- Situações como a do trabalhador que se esforça pouco e a do executivo que toma ações em benefício próprio são chamadas de "problema do agente e do principal"
- O agente é a parte que executa as ações (sem estar sendo perfeitamente monitorado)
- O principal é a parte que contrata o agente para realizar as ações por ele
- Então, o dono da empresa ou acionista são os principais
- Os agentes são os trabalhadores ou os executivos
- O principal pode desenhar contratos com o agente para diminuir o risco moral

### Salário-eficiência

- Uma forma do empregador aumentar o esforço do trabalhador é <strong>pagando um salário acima da média do mercado</strong>
- Ao receber um salário acima da média, o empregador pode reduzir seus custos de monitoramento do trabalhador
- Como a empresa está pagando acima da média, aumenta o custo de oportunidade do trabalhador fazer corpo mole
- Se o trabalhador for pego fazendo corpo mole, ele será demitido
- Como a empresa paga mais que a média, o trabalhador vai se esforçar mais para evitar a demissão
- Assim, um salário mais alto que a média aumenta a produtividade (eficiência do trabalhador, aumentando o lucro da empresa)

### Atrelando a remuneração ao resultado

- Outra forma de aumentar o esforço do trabalhador é alterando a sua remuneração aos seus resultados
- Por exemplo, um vendedor, além do salário fixo, recebe uma porcentagem das vendas que conseguir realizar
- No caso de um executivo, parte de sua remuneração irá depender do lucro da firma
- Mas nem tudo são flores: em geral, a remuneração do executivo está atrelada ao lucro obtido no ano
- Isso gera um incentivo para o executivo priorizar lucros imediatos, e não lucros de longo prazo
- Mas, para obter maiores lucros em um curto prazo, o executivo pode assumir mais riscos do que aqueles que os acionistas gostariam

### Objetivos duplos

- Um outro exemplo são os motoristas de vans
- Motoristas e passageiros têm dois interesses em comum: rapidez no trajeto e fazer uma viagem segura
- Mas, frequentemente, o motorista (o agente) prioriza a rapidez em detrimento da segurança, enquanto o passageiro (o principal) prioriza a segurança em detrimento da rapidez
- Nesse tipo de situação, há dois objetivos envolvidos na relação entre o agente e o principal
- É comum que apenas um dos objetivos seja mensurável, logo, esse objetivo é que será mais "fiscalizado", mas isso prejudicará o outro objetivo
- É mais fácil fiscalizar o cumprimento do horário do que se o motoristaestá dirigindo de forma segura
- O motorista então tenderá a manter o horário, mesmo que isso implique em uma direção perigosa

### Empréstimo bancário

- Um empreendedor toma emprestado de um banco para fazer um investimento em um negócio
- Assim, o empréstimo está usando recursos de terceiros para o investimento
- Caso o investimento não dê certo, o banco executa a garantia dada pelo empresário; mas, normalmente, a garantia é menor que o valor do empréstimo concedido
- Dessa forma, o investidor divide o risco e o prejuízo com o banco
- Dizemos que o empreendedor tem <strong>responsabilidade limitada</strong>
- Devido à responsabilidade limitada, o empreendedor pode tomar mais risco que o desejado pelo banco, o que aumenta a taxa de juros dos empréstimos

### Banco como credor

- Bancos fazem investimentos usando recursos dos acionistas e dos depositantes
- Quanto maior a proporção de recursos de terceiros usados no investimento, mais <strong>alavancado</strong> está o banco
- Em caso de prejuízo nos investimentos, quanto mais alavancado, maior a perda dos depositantes e menor a dos acionistas
- Logo, quanto mais alavancado, maior o incentivo do banco (dos acionistas) em correr riscos
- Essa foi uma das razões que levaram à crise do <i>subprime</i> em 2007

## Seleção adversa

- <strong>Seleção adversa</strong> é um fenômeno que ocorre quando um dos agentes sabe mais sobre suas próprias <strong>características</strong> do que os demais agentes
- Assim, alguns agentes, que chamaremos de "tipo bom", decidem sair do mercado
- No mercado, então, permanece apenas o "tipo ruim"
- O tipo ruim "expulsa" o tipo bom do mercado
-  Ao invés de tipos "bons" e "ruins", é mais comum se usar os termos <strong>"tipo alto"</strong> e <strong>"tipo baixo"</strong>

### O mercado de carros usados sem informação assimétrica

- Suponha que existam dois tipos de carros usado, um bom e outro ruim, e que o vendedor e o comprador sabem que o caro usado bom, para o comprador, vale \\$5000, e para o vendedor, \\$4000
- Como o carro usado bom vale mais para o comprador que para o vendedor, eles irão negociar um preço entre \\$4000 e \\$5000, e o negócio realizado será vantajoso para ambos
- O carro usado ruim, para o comprador, vale \\$1000, e, para o vendedor, vale \\$0 (para simplificar)
- Agora, compradores e vendedores irão negociar algum valor entre \\$0 e \\$1000 e o negócio será vantajoso para ambos
- Os mercados de carros usados bons e ruins estão funcionando perfeitamente

### O mercado de carros usados com informação assimétrica

- Agora suponha que apenas o vendedor sabia se o carro é bom ou ruim, e que existe 50% de probabilidade do carro ser bom
- O comprador não sabe se o carro oferecido é bom ou ruim; por causa dessa incerteza, ele não pagará \\$5000
- Vamos admitir que ele pagará, no máximo, $\frac{1}{2}5000 + \frac{1}{2}1000 = 3000$
- Mas, nesse caso, o dono do carro bom não irá querer vender, já que, para ele, o carro vale \\$4000
- Os donos de carros bons tiram seus carros do mercado, e ficam apenas os ruins!
- A informação assimétrica sobre a qualidade do carro gerou uma seleção adversa: o tipo bom saiu do mercado!

### Resolvendo

- No caso do mercado usado, o problema da assimetria de informação é relativamente fácil de resolver, pois o comprador tem formas de tentar averiguar a qualidade do carro
- Por exemplo, o comprador pode pedir para o vendedor deixar que ele leve o carro em um mecânico de sua confiança
- Mas nem sempre averiguar a qualidade do produto é algo simples
- Pense em um produto eletrônico; muitas vezes, a má qualidade pode demorar meses ou anos para aparecer

## Sinalização e auto-seleção

- <strong>Sinalização</strong> é quando o lado que tem a informação tenta sinalizar que é do tipo bom
- <strong>Auto-seleção</strong> é algo que ocorre quando o lado que não tem a informação oferece várias opções de contratos/produtos para separar o tipo alto do tipo baixo; ao escolher o contrato/produto preferido, o lado que tem a informação revela seu tipo, ou seja, se <strong>auto-seleciona</strong>

## Sinalização

### A cauda do pavão

- O pavão macho tem uma cauda imensa
- Em princípio, a cauda parece ser uma desvantagem pois, além de consumir energia, ainda atrapalha na hora de fugir de um predador
- Então por que a seleção natural gerou essa cauda imensa?
- Biólogos acreditam que a cauda do macho funciona como um <strong>sinal</strong> para a fêmea de que aquele macho tem genes bons
- É uma forma do macho dizer: "olha, tenho genes tão bons que posso ter essa cauda imensa e ainda assim consigo fugir dos predadores"

### Presentes

- A maioria das pessoas não gostaria que seu namorado/namorada lhes dessem dinheiro presente de aniversário
- Porém, a maioria das pessoas não ficaria tão chateada se o mesmo montante de dinheiro lhes fossem dadas de presente pelos próprios pais
- Por que essa diferença de atitude?
- Pais normalmente gostam dos filhos e não precisam sinalizar que gostam deles
- Namorados/namoradas precisam
- Um presente funciona como um sinal, dinheiro não
- Dinheiro é fácil de dar, não exige esforço
- Já um bom presente exige esforço: você precisa saber dos gostos de outra pessoa e, para isso, você precisa prestar atenção na outra pessoa; e as pessoas apreciam isso
- Um bom presente funciona como um sinal de que você realmente gosta e se importa com a outra pessoa e se esforçou para encontrar um presente que se fosse agradar
- Assim, como no caso da cauda do pavão, um sinal somente é um sinal efetivo se for custoso de ser emitido
- Por isso, dinheiro não funciona tão bem como um sinal quanto um presente
- Um bom presente, além do valor monetário, carrega uma informação: "olha, me importo tanto com você que presto atenção e sei do que você gosta"

### Garantias

- Muitas empresas sinalizam a alta qualidade de seu produto oferecendo garantias: "se o produto der defeito em menos de um ano, nós o trocamos gratuitamente"
- Com esse sinal, o comprador consegue sabem que é o tipo bom e quem é o tipo ruim
- Mas, observe que a sinalização só funciona se o tipo ruim não oferecer a garantia também
- Dizemos que não vale a pena para o tipo ruim tentar se passar pelo tipo bom
- Assim como no caso da cauda do pavão, um sinal, para ser efetivo, precisa ser custoso de ser emitido
- Para que apenas o tipo bom ofereça a garantia, <strong>a garantia tem que ser tão custosa para o tipo ruim, que este prefere não fingir que é do tipo bom</strong>
- <strong>O sinal para ser efetivo, precisa ser algo mais custoso para o tipo ruim do que para o tipo bom</strong>
- Se o tipo bom e o tipo ruim conseguem emitir o mesmo sinal, o sinal não é informativo

### Notas altas como um sinal de habilidade

- Um recém-formado, ao procurar um emprego, pode ser de dois tipos: habilidoso (tipo bom) e não habilidoso (tipo ruim)
- A empresa contratante faz entrevistas etc. para tentar separar os dois tipos
- Mas o recém-formado pode tentar sinalizar seu nível de habilidade pelas suas notas, principalmente se for um curso difícil
- As notas altas na faculdade pode sinalizar para o empregador que certo candidato é bom: "se elegonseguiu essas notas altas no curso de engenharia aeroespacial, deve ser porque ele é bom"
- Assim, se o custo de conseguir notas altas é muito alto para o tipo ruim, as notas em cursos ou matérias difíceis carregam informação sobre o tipo do candidato

## Auto-seleção

### Passagens aéreas

- Existem dois tipos de passageiros: os executivos e os turistas
- Companhias aéreas gostariam de cobrar mais dos executivos; mas como fazê-lo se ela não sabe quem é executivo e quem é turista?
- Uma estratégia é cobrar mais de quem compra mais próximo da data da viagem, e menos de quem compra com muita antecendência
- A companhia aérea sabe que, se a pessoa está comprando com antecedência, é mais provável que ela seja turista
- Outra estratégia: se a pessoa for ficar muitos dias no local de destino, maior a probabilidade de ser turista; então, cobra mais de quem for ficar menos dias
