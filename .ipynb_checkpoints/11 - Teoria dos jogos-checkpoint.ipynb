# Teoria dos jogos

- Estuda decisões estratégicas

## Conceitos principais

### Estratégia ótima

Estratégia que maximiza o payoff esperado do jogador

### Jogo

Situação na qual os jogadores (participante) tomam decisões estratégicas que levam em conta as atitudes e respostas um dos outros

### Estratégia

Plano completo de ação para cada situação

### Payoffs

É o objetivo do jogo, o que você deseja maximizar

## Tipos de jogos

### Jogos não cooperativos

Jogos nos quais as negociações e existência de mecanismos que obriguem os cumprimentos de contratos não são possíveis

### Jogos cooperativos

Aqueles nos quais os participantes podem negociar contratos vinculativos de cumprimentos obrigatórios, que permite planejamento conjunto

### Importante

Neste curso, trabaçharemos apenas com jogos não cooperativos

## Exemplo de matriz de payoff

|    |    | Empresa B |     |
| --- | --- | --- | --- |
|    |     | Faz propaganda | Não faz propaganda |
| Empresa A | Faz propaganda | 10,5 | 15,0 |
|    | Não faz propaganda | 6,8 | 10,2 |

### Dois jogadores:
- Empresa A e empresa B

### Duas ações

- Faz ou não faz propaganda

### Previsão do jogo

#### 1ª Etapa: Traçar as estratégias dos jogadores
- Empresa A:
    - A estratégia de A é encontrada quando descobrimos a melhor resposta de A para cada possível ação de B
        - Qual a melhor resposta de A se B faz propaganda?
            - Fazer propaganda
        - Qual a melhor resposta de A se B não faz propaganda?
            - Fazer propaganda
        - Combinando as 2 respostas acima, temos a estratégia ótima de A:
            - Fazer propaganda
- Empresa B:
    - Sabendo que a estratégia de A é sempre a de fazer propaganda, devemos ver qual é a melhor ação de B em resposta à ação de A
        - A estratégia ótima de B é fazer a propaganda também

#### 2ª Etapa: Dadas as estratégias dos jogadores, encontramos a situação em que ambos estão dando as melhores respostas mútuas (Equilíbrio de Nash)

O equilíbrio de Nash do jogo é (Faz propaganda; Faz propaganda), já que as empresas não precisam mais alterar o jogo por conta da resposta da outra, estando assim, as duas satisfeitas, tendo payoffs (10,5)

baixo;esquerda
