# Tarifa em duas partes

- Utilizada em clubes, principalmente
    - O consumidor paga uma taxa de entrada, T, para ser sócio
    - Depois, para cada vez que ele quer praticar alguma atividade dentro do clube, o sócio paga um preço de utilização, P
- Parques também tem essa prática, com um preço de entrada do parque, e outro para utilização de brinquedos

#### Para um parque que é melhor, uma entrada cara e utilizações de brinquedos baratas, ou uma entrada barata e utilizações de brinquedos caras? E quais devem ser estes preços?
Para o parque, é melhor cobrar o custo marginal de utilização em cada volta, e capturar o excedente do consumidor na entrada ao parque.

#### A combinação de preços ótima é cobrar:
- P = CMg cada volta
- T = Excedente do consumidor

# Dois tipos de consumidores

Muitas vezes, as firmas possuem 2 tipos diferentes de consumidores, que possuem curvas de demanda também diferentes, como no exemplo a seguir:

![Gráfico de dois tipos de consumidores](imagens/graficoDoisConsumidores.jpg)

Quando este caso acontece, o monopolista tem que fazer uma escolha no seguinte tradeoff:
- Cobrar, ou o preço de entrada da demanda de consumidores mais baixas, ou a de consumidores mais altas (o que excluirá um dos públicos) e manter o preço marginal como preço de utilização
    - P = CMg
- Cobrar o preço de entrada do grupo mais indisposto e buscar lucro também nas utilizações
    - P > CMg
    - ↓↑ Lucro =↓ Tarifa+ ↑ (P −CMe)Q
