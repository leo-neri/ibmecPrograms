# Monopólio - Parte I

- Monopólios são mercados em que há apenas uma firma ofertando um bem ou serviço, sem substitutos próximos.
- Exemplos: Remédios patenteados, serviços de esgoto, energia elétrica

## Por que existem monopólios?

- Grande poder de mercado
- Barreiras de entrada no mercado

## O comportamento do monopólio

- Lucro econômico sempre positivo (em uma concorrência perfeita, o LE ia para 0)
- Monopolista escolhe ou preço ou quantidade, nunca os dois simultâneos
- Os monopolistas sempre escolhem preço alto e quantidade baixa à preço baixo e quantidade alta
- Sempre o monopolista vai atender a parte elástica da demanda, com |Ed|>1

## Condição de maximização do monopolista

Sempre o monopolista buscará uma situação onde sua receita marginal iguala seu custo marginal

$$Rmg = CMg$$

A regra de preço de um monopolista é:
$$P\;=\;m*CMg(Q),\;onde\; m\;=\;1\;-\;\frac{1}{|Ed|}\;>\;1$$

- CMg(Q) é o custo marginal de produzir Q unidades
- Ed é a elasticidade-preço da demanda
- m é chamado de markup nbsp

A equação do markup nos permite observar que, diferente da concorrência perfeita, onde o preço e o custo marginal se igualavam, sempre o preço do monopólio será maior que o custo marginal, já que o m sempre será maior que 1, e multiplicará a quantidade

## Exercício de fixação

Suponha que a equação da demanda e a equação do custo total do mopolista sejam, respecttivamente:<br><br>$$P = 10 - Q$$<br>$$CT = 3 + Q + 0.5Q²$$

Queremos descobrir qual a quantidade que maximiza o lucro desse monopolista, portanto, precisamos maximixar:

$$L=RT-CT=PQ-CT$$

A receita total será:

$$RT=P*Q=(10-Q)Q=10Q-Q²$$

Daí, a receita marginal será a derivada da receita total em relação a quantidade:<br><br>$$RMg = \frac{dRT}{dQ}=10-2Q$$

Dado o custo total:<br>$$CT = 3+Q+0.5Q²$$

O custo marginal é a derivada do custo total em relação à quantidade:<br><br>
$$CMg=\frac{dCT}{dQ}=1+2*0.5Q=1+Q$$

Igualando a receita marginal ao custo marginal, teremos:<br>
$$RMg=CMg\rightarrow10-2Q=1+Q$$<br>
$$3Q=9$$<br>
$$Q=3$$

Para encontrar o preço, basta substituir Q = 3 na equação de demanda:<br><br>
$$P=10-1=10-3=7$$

O lucro do monopolista será:<br>
$$L=RT-CT=PQ-(3+Q+0.5Q²)$$<br>
$$7*3-(3-3+0.5*3²)$$<br>
$$21-4.5=16.5$$

## Elasticidade-preço da demanda

A fórmula de elasticidade-preço da demanda é:

$$Ed=\frac{P}{D}\frac{1}{\frac{dP}{dQ}}=1-\frac{a}{bQ}$$

No trecho elástico da demanda temos:

$$Ed<-1$$

$$Ed=1-\frac{a}{bQ}<-1\rightarrow2<\frac{a}{bQ}\rightarrow Q<\frac{a}{2b}$$

Portanto, a quantidade que o monopolista escolhe produzir sempre é na primeira metade do triângulo, atrás da curva de RMg.

O monopolista sempre escolhe a parte elástica da demanda, pois ele aumenta o seu lucro conforme aumenta o preço.
