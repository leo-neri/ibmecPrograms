# Venda em pacotes

- A venda em pacotes acontece quando a empresa coloca produtos diferentes em um mesmo pacote e vende todos juntos.
- Exemplo: TV a cabo.
- Ao colocar em um mesmo pacote diversos produtos, as empresas aumentam seu lucro.
- Funciona quando os consumidores têm demandas negativamente correlacionadas pelos produtos do pacote

## Tipos de pacotes

### Puros
- Ocorre quando o consumidor não tem a opção de comprar cada produto separadamente.

### Mistos
- Ocorre quando o consumidor pode comprar cada produto separadamente.
- É melhor quando o CMg de um dos produtos é alto

O que traz mais lucro ao ofertante depende de cada caso.
