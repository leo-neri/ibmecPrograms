# Monopólio - Parte II

## Markup

Representa o quanto a firma vai cobrar acima do custo marginal.

$$m = \frac{1}{1+\frac{1}{Ed}}$$

$$P = m * CMg$$

$$P=\frac{1}{1+\frac{1}{Ed}}*CMg$$

## Exercício de fixação

- Suponha que o custo marginal de produzir um produto seja igual a $9, e que a elasticidade-preço da demanda seja de -2, então:

$$m=\frac{1}{1+\frac{1}{Ed}}= \frac{1}{1+\frac{1}{-2}}=\frac{1}{1-0.5}=\frac{1}{0.5}=2$$

e

$$P=m*CMg=2*9=$18$$

Dizemos que o markup foi de 100(m-1)% = 100(2-1)% = 100%, ou seja o monopolista está colocando um markup de 100% sobre o custo marginal.

O markup também pode acontecer em cima do custo médio, além de que, seu valor sempre é maior que 1.

## Medindo o poder do monopólio - Índice de Lerner

O índice de Lerner, L, é definido como:<br>
$$L=\frac{P-CMg}{P}=1-\frac{CMg}{P}$$

e é uma forma de medir o poder de mercado de um monopólio.

- O índice mede o quão maior é o preço em relação ao custo marginal como uma proporção do preço.<br>
- Quando o preço P é igual ao custo marginal, L=0, ou seja, a firma não tem nenhum poder de monopólio: ela estpa em um mercado de concorrência perfeita.

- Da regra de preço, tínhamos visto que:<br>
$$P=\frac{1}{1+\frac{1}{Ed}}*CMg\rightarrow\frac{P}{CMg}=\frac{1}{1+\frac{1}{Ed}}$$<br>
- Lembrando que a fórmula preço envolve:$$\frac{P}{CMg}=\frac{1}{1+\frac{1}{Ed}},$$ o inverso dela é:$$\frac{CMg}{P}=1+\frac{1}{Ed}.$$
- Combinando com o índice de Lerner:
$$L=\frac{P-CMg}{P}=1-\frac{CMg}{P}=1-(1+\frac{1}{Ed})=-\frac{1}{Ed}$$

- Essa fórmula nos permite ver que o índice de Lerner varia entre 0 e 1:
    - $Ep \rightarrow - \infty,\;L\rightarrow0$: firma sem poder nenhum de monopólio
    - $Ep \rightarrow - 1,\;L\rightarrow1$: firma com poder de monopólio máximo
