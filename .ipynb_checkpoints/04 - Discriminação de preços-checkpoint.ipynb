# Discriminação de preços

Devido ao problema de assimetria de informações, as empresas nunca conseguem saber o preço que cada consumidor está disposto a pagar, e, por conta disso, ela deixa de ganhar. Para diminuir essa perda de ganho, as firmas tentam colocar diferentes preços para diferentes tipos de consumidores.

## Condições para a discriminação de preços existir

- A firma tem algum poder de determinar o preço, ou seja, precisa ter algum poder de monopólio
- Os consumidores podem ser divididos em grupos com diferentes disposições a pagar
- Não é simples haver revenda entre mercados

## Tipos de discriminação

### 1º Grau
- Discriminação perfeita ou personalizada
- O ideal para a firma
- Cobraria um preço diferente para cada pessoa
- Cobraria o máximo que a pessoa está disposta a pagar, obtendo assim, lucro máximo, sem o peso-morto e excedente do consumidor
- Na prática, isso é impossível, devido a assimetria de informação
- A firma não sabe o quanto cada pessoa está disposta a pagar, por isso existem os 2 outros tipos de discriminação

#### Exercício de fixação
- O mercado de voos de avião é formado por 4000 pessoas
- São 2000 executivos que estão dispostos a pagar até R\$550 pela mesma passagem
- São 2000 turistas que estão dispostos a pagar até R\$150 pela mesma passagem
- Suponha que a companhia aérea saiba quem é executivo e quem é turista, e sabe quanto cada um está disposto a pagar
- O custo marginal de cada passagem é de R\$125, constante e igual ao custo médio

#### Perguntas
a) Se a companhia aérea tivesse que vender todas as passagens a um preço único, qual seria o preço? E qual seria o lucro neste caso?<br>
b) Se ela pudesse fazer a discriminação perfeita de preços, qual seria a regra de preços que a companhia faria? Qual seria o lucro neste caso?

#### Respostas
<strong>a)</strong>
A companhia teria que cobrar R\$125 de cada passagem, tendo assim, um lucro de:

```python
print('R$',(150-125)*4000)
```

<strong>b)</strong> A companhia cobraria 550 reais dos executivos, e 150 dos turistas, tendo lucro de:

```python
print('R$',2000*(550+150)-4000*(125))
```

### 2º Grau
- Discriminação por quantidade
- Quanto maior a quantidade, maior o valor total, mas menor o valor unitário
- Todos que compram certa quantidade pagam o mesmo preço
- Sempre se olha para o ganho marginal, não ganho médio
- Um exemplo, são os tamanhos dos milkshakes, onde aumentar do tamanho pequeno para o médio, compensa devido a RMg mais que compensa o acréscimo do CMg

### 3º Grau
- Os consumidores são divididos em grupos, como por exemplo, a nacionalidade
- A firma cobra preços diferentes de cada grupo
- O grande problema das firmas, é separar em grupos, consumidores com características difíceis de se observar. Exemplo: Seguro de carro

## Exercício de fixação

- Em muitos restaurantes os clientes recebem um cartão de desconto<br>
- Por exemplo, depois de 9 refeições, a 10ª é grátis<br>
- Para quem faz as 10 refeições, isso seria o equivalente a um desconto de 10% no preço de cada um dos almoços<br>
<strong>Por que então o restaurante simplesmente não reduz o preço em 10% ao invés de dar o cartão de desconto?</strong>

<strong>R: </strong>
Isso acontece por conta da discriminação de preços que o restaurante faz, tentando ganhar mais dos que estão com maior disposição a pagar (clientes que não vão com muita frequência), e menos dos que estão com menor disposição a pagar (os clientes que mais frequantam o lugar). A discriminação deste caso é uma mistura dos tipos de 2º e 3º grau.
