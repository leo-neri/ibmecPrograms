# Ineficiência do monopólio

## Excedentes do consumidor, produtor e total

- Excedente do produtor (EC) é a diferença do quanto o consumidor estava disposto a pagar, e o quanto ele paga.<br>
- O excedente do produtor (EP), é medido como a diferença entre o preço da venda e o custo marginal de produção.<br>
- O excedente total (ET) é a soma do excedente do produtor e do consumidor.<br>
$$ET = EC + EP$$

Nos monopólios, o consumidor consome menos e paga mais, fazendo o EC ser menor do que na concorrência perfeita. Do lado do produtor, ele ganha excelente devido ao aumento de preço em relação a uma concorrência perfeita. Vale lembrar que a perda do consumidor, é maior do que o agnho do produtor.$$ET\downarrow=EC\downarrow+EP\uparrow$$

## Monopólios naturais

- Monopólios naturais acontecem quan do custo médio de produção é menor quando apenas uma empresa atua no mercado, devido ao enorme custo fixo.
- Nos casos de monopólio de serviços essenciais, é muito comum que subsídios sejam aplicados para que tods tenham acesso, já que empresas não têm interesse de entrar no mercado, porém, isso faz com que a empresa não tenha interesse em diminuir seu custo médio.
