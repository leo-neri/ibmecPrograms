## Maximização do Monopólios

```python
def maximizacaoMonopolio():
    formPreco = input('Insira a fórmula da demanda (ou do preço): P = ').replace('q','Q')
    formCT = input('Insira a fórmula do custo total: CT = ').replace('q','Q')
    P, Q = sp.symbols('P Q')
    preco = sp.sympify(formPreco)
    print('Dados:')
    print(f'\tP = {formPreco}')
    print(f'\tCT = {formCT}')
    print('A receita total (RT), é o preço praticado multiplicado pela quantidade')
    print('\tRT = P * Q')
    print(f'\tRT = ({preco}) * {Q}')
    RT = sp.Mul(preco,Q)
    print('\tRT = '+ str(RT))
    print('A receita marginal (RMg), corresponde à derivada da receita total')
    RMg = sp.diff(RT, Q)
    print(f'\tRMg = dRT/dQ = {str(RMg)}')
    print('Agora que descobrimos a RMg, precisamos descobrir o custo marginal (CMg)')
    CT = sp.sympify(formCT)
    print('\tCT = '+ str(CT))
    CMg = sp.diff(CT, Q)
    print('\tCMg = '+ str(CMg))
    qbugado = str(sp.solveset(sp.Eq(RMg-CMg, 0), Q)).replace('FiniteSet(','').replace(')','')
    try:
        q = float(qbugado)
    except:
        numerador = qbugado.split('/')[0]
        denominador = qbugado.split('/')[1]
        q = float(numerador)/float(denominador)
    print('Tendo a RMg e o CMg, podemos agora descobrir a quantidade')
    print(f'\tRMg = CMg -> {RMg} = {CMg}')
    print(f'\tQ = {q}')
    print('Agora podemos substituir Q na fórmula de demanda, para descobrirmos o preço')
    p = float(str(preco.subs(Q,q) ).replace('FiniteSet(','').replace(')',''))
    print('\tP = ',formPreco.replace('Q',str(q)),' = ',p)
    print('Por fim, vamos descobrir o lucro total (π) da firma monopolista:')
    RTsubstituido = round(RT.subs([(Q,q), (P,p)]),2)
    CTsubstituido = round(CT.subs([(Q,q), (P,p)]),2)
    print(f'\tπ = RT - CT = {RT} - {CT} = {RTsubstituido} - {CTsubstituido} = {RTsubstituido-CTsubstituido}')
    tamEixoX = int(str(sp.solveset(sp.Eq(preco, 0), Q)).replace('FiniteSet(','').replace(')',''))
    fig = plt.figure(figsize=(10,10))
    quantidade = np.linspace(0, tamEixoX, (tamEixoX+1)*10)
    d = eval(formPreco.replace('Q','quantidade'))
    formCMe = '('+formCT+')'+'/Q'
    cme = eval(formCMe.replace('Q','quantidade'))
    cmg = np.array(np.array([CMg for i in range(len(quantidade))]), dtype=float)
    vertical = np.array(np.array([q for i in range(len(quantidade))]), dtype=float)
    rmg = eval(str(RMg).replace('Q','quantidade'))
    ax = fig.add_axes([0,0,1,1])
    ax.set_ylim(ymin=0, ymax = int(float(str(preco.subs(Q,0) ).replace('FiniteSet(','').replace(')',''))))
    ax.plot(quantidade, d, label='D = RMe')
    ax.plot(vertical, d, color="black", ls='--', alpha=0)
    ax.plot(quantidade, cme, label='CMe', color='blue')
    ax.plot(quantidade, cmg, label='CMg', color='red')
    ax.plot(quantidade, rmg, label='RMg', color= 'black')
    encontroCMGeRMG = np.argwhere(np.diff(np.sign(cmg - rmg))).flatten()
    encontroCMEeVertical =np.argwhere(np.diff(np.sign(cme - vertical))).flatten()
    ax.plot(quantidade[encontroCMGeRMG], cme[encontroCMGeRMG], 'ro', color='green',label='CMe em Q*')
    ax.plot(quantidade[encontroCMGeRMG], rmg[encontroCMGeRMG], 'ro', color='black',label='RMg = CMg')
    ax.plot(quantidade[encontroCMGeRMG], d[encontroCMGeRMG], 'ro', color='red', label ='Ponto ótimo de produção')
    ax.plot([quantidade[encontroCMGeRMG],quantidade[encontroCMGeRMG]],[d[encontroCMGeRMG],0], color="black", ls='--') #vertical
    ax.plot([quantidade[encontroCMGeRMG],0],[d[encontroCMGeRMG],d[encontroCMGeRMG]], color="black", ls='--') #horizontal
    ax.fill_between(quantidade, d, color ='grey', alpha=0.01)
    ax.fill_between(quantidade, cmg, d, where=(cmg-d<=0), label='Excedente do consumidor', alpha = 0.5)
    ax.fill_between(quantidade, cme, d, where=(quantidade>=quantidade[encontroCMGeRMG])&(cmg-d<=0), label = 'Peso-morto', color = 'grey')
    ax.fill_between(quantidade, cmg, d[encontroCMGeRMG], where=(quantidade<=quantidade[encontroCMGeRMG])&(cmg-d<-0), label= 'Lucro do monopolista', color = 'yellow')
    ax.legend()
```

## Markup

```python
def markup():
    cmg = float(input('Insira o valor do custo marginal (CMg): '))
    ed = float(input('Insira a elasticidade-preço (Ed) da demanda:'))
    formulaMarkup = '1/(1+(1/Ed))'
    print(f'\nFórmula do markup: {formulaMarkup}')
    print('Substituindo Ed na fórmula: ')
    markupSubstituido = formulaMarkup.replace('Ed',str(ed))
    Ed = sp.symbols('Ed')
    m = sp.symbols('m')
    CMg = sp.symbols('CMg')
    formM = sp.sympify(formulaMarkup)
    markup = formM.subs(Ed,ed)
    print(f'\tm = {formulaMarkup} = {markupSubstituido} = {markup:.2f}')
    print('O preço praticado pelo monopolista é de: ')
    formP = sp.sympify('m*CMg')
    preco = formP.subs([(m,markup), (CMg,cmg)])
    print(f'\tP = m * CMg = {markup:.2f} * {cmg} = ${preco:.2f}')
```

## Índice de Lerner

### Com elasticidade

```python
def lernerElasticidade():
    formulaLerner = '-1/Ed'
    formLerner = sp.sympify(formulaLerner)
    ed = float(input('Insira a elasticidade-preço da demanda: '))
    print(f'Fórmula do índice de Lerner:\n\tL = {formulaLerner}')
    formLernerSubstituido = formulaLerner.replace('Ed',str(ed))
    Ed = sp.symbols('Ed')
    lerner = formLerner.subs(Ed, ed)
    print(f'Substituindo Ed:\n\tL = {formLernerSubstituido} = {lerner:.4f}',)
```

```python
def lernerPeCMg():
    formulaLerner = '1-(CMg/P)'
    formLerner = sp.sympify(formulaLerner)
    cmg = float(input('Insira o custo marginal (CMg) da firma: '))
    p = float(input('Insira o preço (P): '))
    print(f'Fórmula do índice de Lerner:\n\tL = {formulaLerner}')
    formLernerSubstituido = formulaLerner.replace('CMg',str(cmg)).replace('P',str(p))
    CMg = sp.symbols('CMg')
    P = sp.symbols('P')
    lerner = formLerner.subs([(CMg,cmg), (P,p)])
    print(f'Substituindo Ed:\n\tL = {formLernerSubstituido} = {lerner:.4f}',)
```

## Tarifa em duas partes

```python
def tarifaEmDuasPartes():
    P, Q = sp.symbols('P Q')
#     formPreco = '50-Q'
    formPreco = input('Insira a fórmula da demanda (ou do preço): P = ').replace('q','Q')
#     CT = '30+10*Q'
    CT = input('Insira a fórmula do custo total: CT = ')
    CMg = sp.diff(sp.sympify(CT),Q)
    preco = sp.sympify(formPreco)
    RT = sp.Mul(preco,Q)
    RMg = sp.diff(RT, Q)
    qbugado = sp.solve(sp.Eq(RMg-CMg, 0), Q)[0]
    try:
        q = float(qbugado)
    except:
        numerador = qbugado.split('/')[0]
        denominador = qbugado.split('/')[1]
        q = float(numerador)/float(denominador)
    p = float(str(preco.subs(Q,q) ).replace('FiniteSet(','').replace(')',''))
    tamEixoX = int(str(sp.solveset(sp.Eq(preco, 0), Q)).replace('FiniteSet(','').replace(')',''))
    fig = plt.figure(figsize=(10,10))
    quantidade = np.linspace(0, tamEixoX, (tamEixoX+1)*10)
    d = eval(formPreco.replace('Q','quantidade'))
    cmg = np.array(np.array([CMg for i in range(len(quantidade))]), dtype=float)
    vertical = np.array(np.array([q for i in range(len(quantidade))]), dtype=float)
    rmg = eval(str(RMg).replace('Q','quantidade'))
    ax = fig.add_axes([0,0,1,1])
    ax.set_ylim(ymin=0, ymax = int(float(str(preco.subs(Q,0) ).replace('FiniteSet(','').replace(')',''))))
    ax.plot(quantidade, d, label='D = RMe')
    ax.plot(vertical, d, color="black", ls='--', alpha=0)
    ax.plot(quantidade, cmg, label='CMg', color='red')
#     ax.plot(quantidade, rmg, label='RMg', color= 'black')
    encontroCMGeRMG = np.argwhere(np.diff(np.sign(cmg - rmg))).flatten()
#     ax.plot(quantidade[encontroCMGeRMG], rmg[encontroCMGeRMG], 'ro', color='black',label='RMg = CMg')
    ax.fill_between(quantidade, d, color ='grey', alpha=0.01)
    ax.fill_between(quantidade, cmg, d, where=(cmg-d<=0), label='Excedente do consumidor = T', alpha = 0.5)
    ax.legend()
    area1 = np.maximum(d - cmg, 0)[0]
    area2 = np.maximum(d - quantidade, 0)[0]
    print(f'\nCaso a firma queira entrar com um sistema de tarifa de duas partes, com CMg = CMe, o preço de utilização (P) será {CMg}, e o lucro dela será:')
    print(f'T = ({area2}-{CMg}){area1}/2 = {area1*(area2-CMg)/2}')
```

## Equilíbrio de Nash

```python
def equilibrioDeNash():
    jogadasA = input('Quais são as possíveis jogadas de A? Separe por ",": ' ).split(',')
    jogadasB = input('Quais são as possíveis jogadas de B? Separe por ",": ' ).split(',')
    a1b1 = input(f'Insira os dados do campo ({jogadasA[0]},{jogadasB[0]}) separados por ",": ').split(',')
    a1b2 = input(f'Insira os dados do campo ({jogadasA[0]},{jogadasB[1]}) separados por ",": ').split(',')
    a2b1 =  input(f'Insira os dados do campo ({jogadasA[1]},{jogadasB[0]}) separados por ",": ').split(',')
    a2b2 =  input(f'Insira os dados do campo ({jogadasA[1]},{jogadasB[1]}) separados por ",": ').split(',')
    c11a = float(a1b1[0])
    c11b = float(a1b1[1])
    c12a = float(a1b2[0])
    c12b = float(a1b2[1])
    c21a = float(a2b1[0])
    c21b = float(a2b1[1])
    c22a = float(a2b2[0])
    c22b = float(a2b2[1])
    df = pd.DataFrame({'':['','Jogador A',''],' ':['',jogadasA[0],jogadasA[1]],'  ':['','',''],'Jogador B': [jogadasB[0],a1b1,a2b1],'     ': [jogadasB[1],a1b2,a2b2]})
    print()
    print(df)
    if a1b2 == a2b1:
        print(f'\nO jogo possui múltiplos equilíbrios: ({jogadasA[0]},{jogadasB[0]}) e ({jogadasA[1]},{jogadasB[1]})')
        return
    print('\nSe B for: ',jogadasB[0])
    b1 = [c11a,c21a]
    print('\tA ',jogadasA[0],':',c11a)
    print('\tA ',jogadasA[1],':',c21a)
    resultadoA1 = max(b1)
    if resultadoA1 == c21a:
        decisaoA1 = jogadasA[1]
    else:
        decisaoA1 = jogadasA[0]
    print('\tA estratégia do jogador A caso B seja',jogadasB[0],'será:',decisaoA1)
    print('\nSe B for: ',jogadasB[1])
    b2 = [c12a,c22a]
    resultadoA2 = max(b2)
    print('\tA ',jogadasA[0],':',c12a)
    print('\tA ',jogadasA[1],':',c22a)
    payoffA = max(resultadoA1,resultadoA2)
    if resultadoA2 == c12a:
        decisaoA2 = jogadasA[0]
    else:
        decisaoA2 = jogadasA[1]
    print('\tA estratégia do jogador A caso B seja',jogadasB[1],'será:',decisaoA2)
    if decisaoA1 == decisaoA2:
        print('\nA estratégia ótima do jogador A será:',decisaoA1)
    else:
        print('\nEste jogo não possui um equilíbrio de Nash!')
        return
    print('\nCom a estratégia de A definida, vamos ver agora as possibilidades de B para A'+'-'+decisaoA1+':')
    if decisaoA1 == jogadasA[0]:
        a = [c11b,c12b]
        print('\tB ',jogadasB[0],':',c11b)
        print('\tB ',jogadasB[1],':',c12b)
        resultadoB = max(a)
        if resultadoB == c11b:
            payoffB = c11b
            decisaoB = jogadasB[0]
        else:
            payoffB = c12b
            decisaoB = jogadasB[1]
    else:
        a = [c21b,c22b]
        print('\tB ',jogadasB[0],':',c21b)
        print('\tB ',jogadasB[1],':',c22b)
        resultadoB = max(a)
        if resultadoB == c21b:
            payoffB = c21b
            decisaoB = jogadasB[0]
        else:
            payoffB = c22b
            decisaoB = jogadasB[1]
    print('A estratégia que B deve adotar é:',decisaoB)
    print(f'\nO equilíbrio de Nash é ({decisaoA1},{decisaoB})')
```

## Cartel

```python
def cartel():
    strPreco = input('Insira a fórmula do preço: P = ')
#     strPreco = '30 - Q'
    P, Q = sp.symbols('P Q')
    preco = sp.sympify(strPreco)
    custoMarginal = float(input('Insira o custo marginal (CMg): '))
#     custoMarginal = 3.0
    print('A fórmula do lucro dos cartéis é:\nπ = (P-CMg)Q')
    formulaLucro = f'(P - {custoMarginal:.2f})*Q'
    lucroConvertido = sp.simplify(formulaLucro)
    strLucroSubstituido = formulaLucro.replace('P',strPreco)
    formLucroSubstituido = lucroConvertido.subs(P,preco)
    print('Substituindo ela pelos valores dados, temos:')
    print(f'π = {formulaLucro} = {strLucroSubstituido} = {formLucroSubstituido}')
    print('Para descobrirmos a quantidade ideal do mercado, devemos derivar a função do lucro em relação a Q: ')
    lucroDerivado = sp.diff(formLucroSubstituido,Q)
    print(f"{formLucroSubstituido}' = {lucroDerivado}")
    quantidade = float(str(sp.solveset(sp.Eq(lucroDerivado, 0), Q)).replace('FiniteSet(','').replace(')',''))
    print('Resolvendo a derivada:')
    print(f'{lucroDerivado} = 0')
    print(f'Q = {quantidade}')
    print('Precisamos agora obter o preço de mercado:')
    qSubstituido = strPreco.replace('Q',str(quantidade))
    mercado = sp.sympify(qSubstituido).subs(Q,quantidade)
    print(f'P = {strPreco} = {qSubstituido} = {mercado:.2f}')
    print('Por fim, vamos descobrir o lucro conjunto do cartel:')
    formulaFinal = formulaLucro.replace('P',str(mercado)).replace('Q',str(quantidade))
    lucroCartel = (mercado-custoMarginal)*quantidade
    print(f'π = (P - CMg)*Q = {formulaFinal} = {lucroCartel}')
    print('Supondo que elas dividem o lucro:')
    print(f'π1 = π2 = {lucroCartel/2}')
```

```python
def bertrandDiferenciados():
    P1, Q1, P2, Q2 = sp.symbols('P1 Q1 P2 Q2')
    escolha = input('Deseja inserir os lucros [1] ou as curvas de demanda? [2]: ')
    if escolha == '1':
        lucro1 = input('Insira a fórmula dos lucros da empresa 1: π1 = ')
        lucro2 = input('Insira a fórmula dos lucros da empresa 2: π2 = ')
        formLucro1 = sp.sympify(lucro1)
        formLucro2 = sp.sympify(lucro2)
        custoTotal1 = lucro1.split()[-1]
        custoTotal2 = lucro2.split()[-1]
        ct1Calculo = sp.sympify(custoTotal1)
        ct2Calculo = sp.sympify(custoTotal2)
    else:
        demanda1 = input('Insira a curva de demanda da empresa 1: Q1 =')
        demanda2 = input('Insira a curva de demanda da empresa 2: Q2 =')
        custoTotal1 = input('Insira a fórmula do custo total da empresa 1: CT1 =')
        custoTotal2 = input('Insira a fórmula do custo total da empresa 2: CT1 =')
        formReceita1 = 'P1*Q1'
        formReceita2 = 'P2*Q2'
        d1Calculo = sp.sympify(demanda1)
        d2Calculo = sp.sympify(demanda2)
        ct1Calculo = sp.sympify(custoTotal1)
        ct2Calculo = sp.sympify(custoTotal2)
        formDemanda1 = sp.sympify(formReceita1).subs(Q1,demanda1)
        formDemanda2 = sp.sympify(formReceita2).subs(Q2,demanda2)
        print(formDemanda1)
        print('Considerando a fórmula de receita como P*Q, temos:')
        print(f'R1 = {formReceita1} = P1*({demanda1})')
        print(f'R2 = {formReceita2} = P1*({demanda2})')
        print('Vamos agora calcular os lucros de cada firma:')
        formLucro1 = sp.Mul(P1,d1Calculo) - ct1Calculo
        formLucro2 = sp.Mul(P2,d2Calculo) - ct2Calculo
        print(f'π1 = R1 - C1 = {formLucro1}')
        print(f'π2 = R2 - C2 = {formLucro2}')
    custoMarginal = sp.diff(ct1Calculo,Q1)
    print('Para descobrimos as funções de maximização, precisamos derivar as fórmulas de lucro e igualar a 0:')
    formMax1 = sp.diff(formLucro1,P1)
    formMax2 = sp.diff(formLucro2,P2)
    print(f"π1' = {formMax1}")
    print(f"π2' = {formMax2}")
    resultadoP1 = str(sp.solveset(formMax1, P1)).replace('FiniteSet(','').replace(')','')
    resultadoP2 = str(sp.solveset(formMax2, P2)).replace('FiniteSet(','').replace(')','')
    print('As funções de reação das firmas são:')
    print(f'P1 = {resultadoP1}')
    print(f'P2 = {resultadoP2}')
    print('Para encontrar as soluções, vamos substituir as incógnitas entre as equações:')
    resultado1Substituido = resultadoP1.replace('P2','('+resultadoP2+')')
    resultado2Substituido = resultadoP2.replace('P1','('+resultadoP1+')')
    solucao1 = str(sp.solve(sp.Eq(sp.sympify(resultadoP1).subs(P2,sp.sympify(resultadoP2)),P1),P1)).replace('[','').replace(']','')
    solucao2 = str(sp.solve(sp.Eq(sp.sympify(resultadoP2).subs(P1,sp.sympify(resultadoP1)),P2),P2)).replace('[','').replace(']','')
    print(f'P1 = {resultado1Substituido} = {solucao1}')
    print(f'P2 = {resultado2Substituido} = {solucao2}')
    if float(solucao1) > float(solucao2):
        primeiraParte = 'P1 > P2 = '
    elif float(solucao1) == float(solucao2):
        primeiraParte = 'P1 = P2 = '
    else:
        primeiraParte = 'P1 < P2 = '
    if float(solucao2) > custoMarginal:
        segundaParte = ' > CMg = '
    elif float(solucao2) == custoMarginal:
        segundaParte = ' == CMg = '
    else:
        segundaParte = ' < CMg = '
    print(primeiraParte+solucao2+segundaParte+str(custoMarginal))
    print(f'O equilíbrio de Nash é ambas as firmas ganharem {solucao1}.')
    # Gráficos
    x = np.linspace(0,101,1000)
    y = eval(resultadoP1.replace('P2','x'))
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_axes([0,0,1,1])
    equilibrioDeNash = np.argwhere(np.diff(np.sign(x-y))).flatten()
    ax.plot(x[equilibrioDeNash], y[equilibrioDeNash], 'ro', color='green',label='Equilíbrio de Nash')
    ax.plot(x,y, label = 'P1')
    ax.plot(y, x, label = 'P2')
    ax.set_xlabel('P1')
    ax.set_ylabel('P2')
    ax.legend()
```

## Cournot

```python
def cournot():
    Q1, Q2, Q, P = sp.symbols('Q1 Q2 Q P')
#     formCusto1 = '40*Q1'
    formCusto1 = input('Insira o custo de criar capacidade produtiva da empresa 1: CT1(Q1) = ')
#     formCusto2 = '14*Q2'
    formCusto2 = input('Insira o custo de criar capacidade produtiva da empresa 2: CT2(Q2) = ')
#     formPreco = '400-2*Q'
    formPreco = input('Insira a função do preço: P =')
    print('Dados:')
    print('Custo firma I: CT =',formCusto1)
    print('Custo firma II: CT =',formCusto2)
    print('Preço: P =',formPreco)
    preco = sp.sympify(formPreco)
    print('\nA fórmula da quantidade no modelo de Cournot é:')
    print(f'\tQ =  Q1 + Q2')
    print('Substituindo na fórmula do preço (P):')
    precoSubstituido = preco.subs(Q,Q1 + Q2)
    print(f'\tP = {precoSubstituido}')
    print('\nVamos agora descobrir o lucro da firma I:')
    formLucro1 = 'P*Q1 - '+ formCusto1
    lucro1Substituido = sp.sympify(formLucro1).subs(P,precoSubstituido)
    lucro1Expandido = sp.expand(lucro1Substituido)
    lucro1Simplificado = sp.simplify(lucro1Substituido)
    print(f'\tπ1 = {formLucro1} = {lucro1Substituido} = {lucro1Expandido} = {lucro1Simplificado}')
    print('Para maximizar o lucro da firma I, precisamos derivar o lucro em relação a Q1, e depois, igualar a 0:')
    lucro1Derivado = sp.diff(lucro1Simplificado,Q1)
    lucro1Resolvido = sp.solve(sp.Eq(lucro1Derivado,0),Q1)
    lucro1ResolvidoStr = str(lucro1Resolvido).replace('[','').replace(']','')
    print(f"\tπ1' = {lucro1Derivado}")
    print(f'\tQ1 = {lucro1Derivado} = 0')
    print('Resolvendo a equação, obtemos a função de reação da primeira firma:')
    print(f"\tQ1 = {lucro1ResolvidoStr}")
    
    print('\nVamos agora descobrir o lucro da firma II:')
    formLucro2 = 'P*Q2 - '+ formCusto2
    lucro2Substituido = sp.sympify(formLucro2).subs(P,precoSubstituido)
    lucro2Expandido = sp.expand(lucro2Substituido)
    lucro2Simplificado = sp.simplify(lucro2Substituido)
    print(f'\tπ2 = {formLucro2} = {lucro2Substituido} = {lucro2Expandido} = {lucro2Simplificado}')
    print('Para maximizar o lucro da firma II, precisamos derivar o lucro em relação a Q2, e depois, igualar a 0:')
    lucro2Derivado = sp.diff(lucro2Simplificado,Q2)
    lucro2Resolvido = sp.solve(sp.Eq(lucro2Derivado,0),Q2)
    lucro2ResolvidoStr = str(lucro2Resolvido).replace('[','').replace(']','')
    print(f"\tπ2' = {lucro2Derivado}")
    print(f'\tQ2 = {lucro2Derivado} = 0')
    print('Resolvendo a equação, obtemos a função de reação da segunda firma:')
    print(f"\tQ2 = {lucro2ResolvidoStr}")
    
    print('\nVamos agora reescrever a função de reação da firma I em função de Q2:')
    reacao1Reescrita = sp.solve(sp.Eq(sp.sympify(lucro1ResolvidoStr),Q1),Q2)
    reacao1ReescritaStr = str(reacao1Reescrita).replace('[','').replace(']','')
    print(f'\tQ1 = {lucro1ResolvidoStr}')
    print(f'\tQ2 = {reacao1ReescritaStr}')
    print('Vamos agora igualar esta última equação à função de reação da firma II para descobrirmos o valor de Q1:')
    print(f'\t{lucro2ResolvidoStr} = {reacao1ReescritaStr}')
    q1 = float("%.2f" % float(sp.solve(sp.Eq(sp.sympify(reacao1ReescritaStr),sp.sympify(lucro2ResolvidoStr)),Q1)[0]))
    print(f'\tQ1 = {q1}')
    
    print('\nVamos agora reescrever a função de reação da firma II em função de Q1:')
    reacao2Reescrita = sp.solve(sp.Eq(sp.sympify(lucro2ResolvidoStr),Q2),Q1)
    reacao2ReescritaStr = str(reacao2Reescrita).replace('[','').replace(']','')
    print(f'\tQ2 = {lucro2ResolvidoStr}')
    print(f'\tQ1 = {reacao2ReescritaStr}')
    print('Vamos agora igualar esta última equação à função de reação da firma I para descobrirmos o valor de Q2:')
    print(f'\t{lucro1ResolvidoStr} = {reacao2ReescritaStr}')
    q2 = float("%.2f" % float(sp.solve(sp.Eq(sp.sympify(reacao2ReescritaStr),sp.sympify(lucro1ResolvidoStr)),Q2)[0]))
    print(f'\tQ2 = {q2}')
    
    print(f'\nDescobrindo a capacidade produtiva:')
    capacidadeProdutiva = "%.2f" % float(q1+q2)
    print(f'\tQ = Q1 + Q2 = {q1} + {q2} = {capacidadeProdutiva}')
    
    formPrecoSubstituido = formPreco.replace('Q',str(capacidadeProdutiva))
    print(f'Descobrindo o preço praticado:')
    precoPraticado = "%.2f" % float(preco.subs(Q,q1+q2))
    print(f'\tP = {formPreco} = {formPrecoSubstituido} = {precoPraticado}')
    
    print('Por último, vamos ver os lucros de cada firma:')
    custo1Substituido = formCusto1.replace('Q1',str(q1))
    custo2Substituido = formCusto2.replace('Q2',str(q2))
    lucroLider = float("%.2f" % float(preco.subs(Q,q1+q2)*q1-sp.sympify(custo1Substituido)))
    lucroSeguidora = float("%.2f" % float(preco.subs(Q,q2+q1)*q2-sp.sympify(custo2Substituido)))
    print(f'\tπ1 = P*Q1 - CT1 = {precoPraticado}*{q1} - {custo1Substituido} = {lucroLider}')
    print(f'\tπ2 = P*Q2 - CT2 = {precoPraticado}*{q2} - {custo2Substituido} = {lucroSeguidora}')
    
    #Gráfico
    x = np.linspace(0,101,1000)
    y = eval(lucro2ResolvidoStr.replace('Q1','x'))
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_axes([0,0,1,1])
    equilibrioDeNash = np.argwhere(np.diff(np.sign(x-y))).flatten()
    ax.plot(x[equilibrioDeNash], y[equilibrioDeNash], 'ro', color='green',label='Equilíbrio de Nash/Cournot')
    ax.plot(x,y, label = 'Firma I')
    ax.plot(y, x, label = 'Firma II')
    ax.set_xlabel('Q2')
    ax.set_ylabel('Q1')
    ax.set_ylim(ymin=0, ymax = 100)
    ax.set_xlim(xmin=0, xmax = 100)
    ax.legend()
```

## Stackelberg

```python
def stackelberg():
    Q1, Q2, Q, P = sp.symbols('Q1 Q2 Q P')
    formLider = input('Insira o custo de criar capacidade produtiva da empresa líder: C1(Q1) = ')
#     formLider = '40*Q1'
    formSeguidora = input('Insira o custo de criar capacidade produtiva da empresa seguidora: C2(Q2) = ')
#     formSeguidora = '32*Q2'
    formPreco = input('Insira a função do preço: P = ')
#     formPreco = '400-2*Q'
    print('Dados:')
    print('Custo firma I: CT =',formLider)
    print('Custo firma II: CT =',formSeguidora)
    print('Preço: P =',formPreco)
    preco = sp.sympify(formPreco)
    precoSubstituido = preco.subs(Q,Q1 + Q2)
    print('\nVamos descobrir o lucro da firma seguidora:')
    formSeguidora = 'P*Q2 - '+ formSeguidora
    seguidoraSubstituido = sp.sympify(formSeguidora).subs(P,precoSubstituido)
    seguidoraExpandido = sp.expand(seguidoraSubstituido)
    seguidoraSimplificado = sp.simplify(seguidoraSubstituido)
    print(f'\tπ2 = {formSeguidora} = {seguidoraSubstituido} = {seguidoraExpandido} = {seguidoraSimplificado}')
    print('Para maximizar o lucro da firma seguidora, precisamos derivar o lucro em relação a Q2, e depois, igualar a 0:')
    seguidoraDerivado = sp.diff(seguidoraSimplificado,Q2)
    seguidoraResolvido = sp.solve(sp.Eq(seguidoraDerivado,0),Q2)
    seguidoraResolvidoStr = str(seguidoraResolvido).replace('[','').replace(']','')
    print(f"\tπ2' = {seguidoraDerivado}")
    print(f'\t{seguidoraDerivado} = 0')
    print('Resolvendo a equação, obtemos a função de reação da firma seguidora:')
    print(f"\tQ2 = {seguidoraResolvidoStr}")
    
    print('\nVamos agora descobrir o lucro da firma líder:')
    formLider = 'P*Q1 - '+ formLider
    liderSubstituido = sp.sympify(formLider).subs(P,precoSubstituido)
    liderExpandido = sp.expand(liderSubstituido)
    print(f'\tπ1 = {formLider} = {liderSubstituido} = {liderExpandido}')
    
    print('\nVamos agora substituir Q2, obtido na função de reação da empresa seguidora, na fórmula de maximização de lucro da empresa líder:')
    liderExpandidoSubstituido = sp.expand(liderSubstituido.subs(Q2,sp.sympify(seguidoraResolvidoStr)))
    print(f'\tπ1 = {liderExpandido} = {liderExpandidoSubstituido} = {sp.simplify(liderExpandidoSubstituido)}')
    
    print('\nPara acharmos o lucro máximo da firma líder, primeiro vamos derivar em função de Q1 a fórmula de lucro dela:')
    liderDerivada = sp.diff(liderExpandidoSubstituido,Q1)
    print(f"\tπ1' = ({sp.simplify(liderExpandidoSubstituido)})' = {liderDerivada}'")
    print('Achada a derivada, vamos igualar ela a 0 e achar o valor de Q1:')
    q1 = float("%.2f" % float(sp.solve(sp.Eq(liderDerivada,0),Q1)[0]))
    print(f'\t{liderDerivada} = 0')
    print(f'\tQ1 = {q1}')
    print('Substituindo Q1 na função de reação da firma seguidora:')
    seguidoraResolvidoStrSubstituido = seguidoraResolvidoStr.replace('Q1',str(q1))
    q2 = float("%.2f" % float(sp.sympify(seguidoraResolvidoStrSubstituido)))
    print(f'\tQ2 = {seguidoraResolvidoStr} = {seguidoraResolvidoStrSubstituido} = {q2}')
    print('Vamos agora descobrir o preço, considerando que Q = Q1 + Q2')
    formPrecoSubstituido = formPreco.replace('Q','('+str(q1)+'+'+str(q2)+')')
    p = float("%.2f" % float(sp.sympify(formPrecoSubstituido)))
    print(f'\tP = {formPreco} = {formPrecoSubstituido} = {p}')
    print('Agora podemos descobrir o lucro de cada firma:')
    liderSubstituido = formLider.replace('Q1',str(q1)).replace('P',str(p))
    SeguidoraSubstituido = formSeguidora.replace('Q2',str(q2)).replace('P',str(p))
    lucroLider = "%.2f" % float(sp.sympify(liderSubstituido))
    lucroSeguidora = "%.2f" % (sp.sympify(SeguidoraSubstituido))
    print(f'\tπ1 = P*Q1 - CT1 = {liderSubstituido} = {lucroLider}')
    print(f'\tπ2 = P*Q2 - CT2 = {SeguidoraSubstituido} = {lucroSeguidora}')
    
    #Gráfico
    x = np.linspace(0,101,1000)
    y = eval(seguidoraResolvidoStr.replace('Q1','x'))
    fig = plt.figure(figsize=(10,10))
    ax = fig.add_axes([0,0,1,1])
    equilibrioDeNash = np.argwhere(np.diff(np.sign(x-y))).flatten()
    ax.plot(x[equilibrioDeNash], y[equilibrioDeNash], 'ro', color='green',label='Equilíbrio de Nash')
    ax.plot(x,y, label = 'Líder')
    ax.plot(y, x, label = 'Seguidora')
    ax.set_xlabel('Q2')
    ax.set_ylabel('Q1')
    ax.set_ylim(ymin=0, ymax = 100)
    ax.set_xlim(xmin=0, xmax = 100)
    ax.legend()
```
