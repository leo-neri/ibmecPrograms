# Oligopólios

- Mercado caraterizado pela existência de poucas empresas que produzem produtos iguais
- Normalmente existem barreiras à entrada de novas firmas no mercado

## Classificação de oligopólios

- Produtos
    - Bens homogêneos
    - Bens diferenciados
- Competição
    - Preço
    - Quantidade
- Sequência temporal
    - Simultaneidade
    - Uma empresa líder decide primeiro e as demais seguem

## Bertrand

- Apenas duas empresas
- Bens homogêneos ou diferenciados
- Competem em preço
- Decidem simultaneamente suas ações
- No duopólio de Bertrand, o equilíbrio de Nash é aquela situação em que cada empresa está satisfeita com o preço que escolheu
    - Nenhuma delas aumenta seu lucro se alterar sozinha o seu preço
    - Exceção aos cartéis

### Bens homogêneos

#### O mercado

- Suponha duas empresas idênticas, com custo marginal de produção constante e igual a R\$3: CMg = 3
- Suponha, para simplificar, que se os preços das duas empresas fossem iguais, metade das pessoas comprariam da empresa 1 e a outra metade compraria da empresa 2
- Note também que qualquer que seja o preço que as empresas escolherem, é necessário que ele seja maior ou igual ao custo marginal:<br> $$P1 >=3\;\;e\;\;P2 >= 3$$
- No modelo de Bertrando, sempre o EN é ambas as empresas praticarem o preço igual ao custo marginal, assim como na concocrrência perfeita, sendo assim, o lucro econômico é 0

#### Formas de reduzir a competição

Para que o lucro seja maior que 0, as empresas participantes de oligopólios tentam reduzir a competição, e algumas formas de fazer isso são:
- Formação de cartéis
    - Firmas combinam manter o preço elevado, punindo quem reduz o preço
- Diferenciação de produtos
    - Para ganhar poder de mercado
- Limitação da capacidade de produção
    - Com capacidade produtiva limitada, uma firma não consegue alimentar toda a demanda

#### Cartéis

- Um cartel ocorre quando um conjunto de firmas entra em acordo sobre o preço que todas irão cobrar e/ou a quantidade que irão produzir
- A ideia é que elas operem como se fosse uma única empresa monopolista

### Bens diferenciados

- Com a diferenciação de produtos, as firmas conseguem cobrar preços acima do custo marginal

## Cournot

- O modelo de Cournot é uma forma simples de representar um jogo em dois estágios:
    1. Primeiro: Firmas que produzem bens homogêneos decidem simultaneamente a capacidade de produção (o tamanho da fábrica)
    2. Segundo: Decidida a capacidade de produção, as firmas competem via preços, como no modelo de Bertrand
- A escolha de uma capacidade produtiva não muito alta no primeiro estágio permitirá às firmas terem lucros econômicos positivos no segundo estágio

## Stackelberg

- O modelo de Stackelberg é uma variação do modelo de Cournot
- Agora, há uma firma líder que escolhe sua capacidade de produção primeiro
